```python
from pandas import DataFrame
from json import loads
import numpy as np
```

```python
path = 'url-shortner/usagov_bitly_data2012-03-16-1331923249.txt'
file = open(path, mode='r')
lines = []
for line in file:
    obj = loads(line)
    lines.append(obj)
```

```python
data = DataFrame(lines)
data
```

```python
s1 = data['tz']
s2 = s1.unique()
s2
```

```python
s3 = s1[(s1 == '') | (s1.isnull())]
s3
```

```python
s4 = s1.fillna('Missing')
s4[s4 == '']
```

```python
s4[s4 == ''] = 'Missing'
```

```python
s4[(s4 == '') | (s4.isnull())]
```

```python
s4.head(20)
```

```python
data['tz'] = s4
data.head(20)
```

```python
d1 = data['tz'].value_counts().head(10)
d1
```

```python
d1.plot(kind='bar')
```

```python
d2 = np.where(data['a'].str.contains('Windows'), 'Windows', 'Non-windows')
d2
```

```python
data['os'] = d2
```

```python
data[['a', 'os']].head(20)
```

```python
d3 = data[['tz', 'os']]
d3
```

```python
d3['total'] = 1
d3
```

```python
d4 = d3.pivot_table('total', aggfunc='sum', index='tz', columns='os')
d4
```

```python
d5 = d4.loc[d1.index]
d6 = d5.fillna(0)
d6
```

```python
d6.plot(kind='bar')
```
